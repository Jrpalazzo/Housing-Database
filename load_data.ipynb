```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
# Create a connection to the database. File will be created if it doesn't already exist

conn = sqlite3.connect('./Census.db')

curs = conn.cursor()

#always keep foreign_keys on when you are changing data
curs.execute("PRAGMA foreign_keys = off;")
```

```python
loc = pd.read_csv('./location_data.csv')
```

```python
hsg = pd.read_csv('./housing_data.csv')
```

```python
#change column names
emp = pd.read_csv('./employment_data.csv')
emp.columns = (['ID','ID2','total_estimate', 'total_margin', 'labor_estimate', 'labor_margin','ratio_estimate', 'ratio_margin', 'unemployment_estimate','unemployment_margin','year', 'county', 'state'])
emp.head(3)
```

```python
#There are NAN rows in the data set, drop them
emp.dropna()
```

```python
#change column names
inc = pd.read_csv('./income_data.csv')
inc.columns = (['ID', 'ID2', 'median_estimate', 'mean_estimate', 'year', 'county', 'state'])

inc.head(3)
```

# Load the Data

```python

# Load data in same order as tables were created

#use int(x[0]) to make sure you are adding a regular integer.
#pandas.sql can be used to load data

# 1) tlocation
for w in loc.values:
    curs.execute("""INSERT INTO tLocation VALUES(
                    ?,?,?,?
                    );""", ( w[0], w[1], w[2], w[3] ) )

# 2) tHousing

for x in hsg.values:
    curs.execute("""INSERT INTO tHousing VALUES(
                    ?,?,?,?
                    );""", ( x[0], x[1], x[2], x[3] ) )
    
# 3) tEmployment

for y in emp.values:
    curs.execute("""INSERT INTO tEmployment VALUES(
                    ?,?,?,?,?,?,?,?,?,?,?
                    );""", ( y[0], y[1], y[2], y[3], y[4], y[5], y[6], y[7], y[8], y[9], y[10]) )

# 4) tIncome

for z in inc.values:
    curs.execute("""INSERT INTO tIncome VALUES(
                    ?,?,?,?,?
                    );""", ( z[0], z[1], z[2], z[3], z[4]) )
conn.commit()
```
