```python
#//////////////////////////Explore the Income Files/////////////////////////////////////////////////////
```

```python
#we need to create a new year column
df_income0 = pd.read_csv('./ACS_15_5YR_S1901_with_ann.csv',encoding = "ISO-8859-1",skiprows=1)
df_income0 = df_income0.assign(year = ['2015']*len(df_income0.index))

df_income1 = pd.read_csv('./ACS_16_5YR_S1901_with_ann.csv',encoding = "ISO-8859-1",skiprows=1)
df_income1 = df_income1.assign(year = ['2016']*len(df_income1.index))

df_income2 = pd.read_csv('./ACS_17_5YR_S1901_with_ann.csv',encoding = "ISO-8859-1",skiprows=1)
df_income2 = df_income2.assign(year = ['2017']*len(df_income2.index))
```

```python
df_income = pd.concat([df_income0, df_income1, df_income2])
list(df_income)
```

```python
df_income.info()
```

```python
list(df_income)
```

```python
#df.drop(list(df.filter(regex = 'Test')), axis = 1, inplace = True)
#get all the columns that we don't need and drop them by finding a part of the word with 'like'
drop_income = df_income.drop(df_income.filter(like='Families',axis=1).columns,axis=1)
drop_income2 = drop_income.drop(drop_income.filter(like='Married',axis=1).columns,axis=1)
drop_income3 = drop_income2.drop(drop_income2.filter(like='Nonfamily',axis=1).columns,axis=1)
drop_income4 = drop_income3.drop(drop_income3.iloc[:, 3:23], axis=1)
#data_long_income = (pd.melt(drop_income5, id_vars = ['Geography', 'year']
#                     , var_name = 'housing', value_name = 'housing_data' ))
```

```python
drop_income4.head(3)
```

```python
#split state and county 
#drop columns we don't need
drop_income4 = drop_income4.drop(columns = ['Households; Estimate; $200,000 or more', 'Households; Margin of Error; $200,000 or more'
                                            ,'Households; Margin of Error; Median income (dollars)', 'Households; Margin of Error; Mean income (dollars)'])
drop_income4[['county', 'state']] = (drop_income4['Geography'].str.split(", ", 1, expand=True))            
data_income = drop_income4.drop(columns = ['Geography'])
data_income.head(3)  
```

```python
data_income.to_csv('income_data.csv', index=False) #store data set to a new csv file 
```
